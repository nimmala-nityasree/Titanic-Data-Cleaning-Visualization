```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
import matplotlib.pyplot as plt

df['Survived'].value_counts().plot(kind='bar')

plt.title("Survival Count")
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.hist(df['Age'], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

Observations

1. More passengers did not survive than survived.

2. The dataset contains more male passengers than female passengers.

3. Most passengers were between 20 and 40 years old.

4. Missing values in the Age column were handled using the mean age.
